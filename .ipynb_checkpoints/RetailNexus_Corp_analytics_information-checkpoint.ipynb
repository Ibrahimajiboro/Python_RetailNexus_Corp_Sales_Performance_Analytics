Use **IQR + Log Transform** on `Sales`, `Profit`, and `Discount`:

```python
import numpy as np

# Log Transform
df['Sales_log'] = np.log1p(df['Sales'])

# IQR Removal
Q1 = df['Profit'].quantile(0.25)
Q3 = df['Profit'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['Profit'] >= Q1 - 1.5*IQR) & (df['Profit'] <= Q3 + 1.5*IQR)]
```
